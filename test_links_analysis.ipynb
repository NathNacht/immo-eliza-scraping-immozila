```python
import json
```

```python
with open ("list_of_links.json") as p:
    list_of_links = json.load(p)
    
```

```python
print (len(list_of_links))
```

```python
splitted_links = []
for item in list_of_links:
    splitted_link = item.split("/")
    splitted_link.remove('https:')
    splitted_link.remove('')
    splitted_link.remove('www.immoweb.be')
    splitted_link.remove('nl')
    splitted_link.remove('zoekertje')
    splitted_links.append(splitted_link)

with open ("splitted_links.json", "w", encoding="utf-8") as f:
    json.dump(splitted_links, f)
```

```python
import pandas as pd
```

```python
places =[ i[-3] for i in splitted_links]
print(len(places))
places_deduplicated = list(set(places))
print(len(places_deduplicated))
codes = [i[-2] for i in splitted_links]
print(len(codes))

places_count = []
for code in codes:
    count = codes.count(code)
    places_count.append(count)

print(len(places_count))
print(places_count)
```

```python
places_dict = {"place": places, "postcode": codes, "count": places_count}
df = pd.DataFrame(places_dict)

df = df.sort_values("count", ascending=False)

print(df)

df.to_excel("cities_with_codes.xlsx")
```
